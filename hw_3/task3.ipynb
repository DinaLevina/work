```python
from google.colab import drive
import pandas as pd

#drive.mount('/content/gdrive')

#%cd gdrive/MyDrive/

df=pd.read_csv("train.csv")
```

Пропуски данных:

```python
df.isna().sum()
```

Выживаемость по классам:

```python
scl=df.groupby("Pclass")["Survived"].sum()
print(scl)
```

Всего людей по классам:

```python
pcl=df['Pclass'].value_counts()
print(pcl)
cls=[]
cls.append(pcl.iloc[1])
cls.append(pcl.iloc[2])
cls.append(pcl.iloc[0])
print(cls)
```

Процент выживаемости по калссам:

```python
for i in range(1,4):
  print("Class",i,(scl[i]/cls[i-1])*100,"%")
```

Самое популярное мужское имя:

```python
name=df[df['Sex']=='male']
name=name['Name']
nlist=[]
for i in range(len(name)):
  n=name.iloc[i]
  nlist=nlist+n.split(' ')
mn=pd.DataFrame(nlist,columns=['MenNames'])
mn['MenNames'].value_counts()
```
