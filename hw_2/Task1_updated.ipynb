```python
class Chocolate:
    def __init__(self,price:float,addon:str=None):
        if price > 0:
            self.price=price
        else:
            raise ValueError('Цена должна быть больше 0!')
        self.addon=addon

    def show_my_chocolate(self):
        print(f"Шоколадный батончик, добавка: {self.addon}")

    def __str__(self):
        return f'Шоколадный батончик, <{self.price}> рублей'


bar1=Chocolate(200,'Кокос')
print(bar1)
bar1.show_my_chocolate()
bar2=Chocolate(5)
bar3=Chocolate(300,'Орехи')
print(bar3)
print(bar2)
bar4=Chocolate(-10)
```
