```python
import pandas as pd
import requests
api_base_url = 'https://www.anapioficeandfire.com/api'
```

Получаем список книг и сохраняем в DataFrame

```python
book_list = requests.get(f'{api_base_url}/books',).json()
result=[]
for i in range(len(book_list)):
  book={}
  book['authors']=book_list[i]['authors']
  book['name']=book_list[i]['name']
  result.append(book)
pd.set_option('display.max_colwidth', 400)
data = pd.DataFrame(result)
data.head()
```

Получаем список домов и сохраняем в DataFrame

```python
house_list = requests.get(f'{api_base_url}/houses',).json()
result=[]
for i in range(len(house_list)):
  house={}
  house['region']=house_list[i]['region']
  house['name']=house_list[i]['name']
  result.append(house)
pd.set_option('display.max_colwidth', 400)
data = pd.DataFrame(result)
data.head()
```

Получаем список домов с девизом и сохраняем в DataFrame

```python
house_list = requests.get(f'{api_base_url}/houses?hasTitles=true',).json()
result=[]
for i in range(len(house_list)):
  house={}
  house['titles']=house_list[i]['titles']
  house['name']=house_list[i]['name']
  result.append(house)
pd.set_option('display.max_colwidth', 400)
data = pd.DataFrame(result)
data.head()
```
