Получение ссылок

```python
import requests
from bs4 import BeautifulSoup
base_url = 'https://lifehacker.ru/topics/technology'
#response = requests.get('https://lifehacker.ru/topics/technology/')
parsed_urls=[]
for page_num in range(1, 11):
  if page_num==1:
    url=f'{base_url}'
  else:
    url=f'{base_url}/?page={page_num}'
  response = requests.get(url)
  soup = BeautifulSoup(response.text, 'lxml')
  raw_items=soup.find_all('a', class_='lh-small-article-card__link')
  parsed_urls += [item.get('href') for item in raw_items]
parsed_urls
```

Получение текста конкретной статьи

```python
response = requests.get('https://lifehacker.ru/robo-kon-corleo/')
soup = BeautifulSoup(response.text, 'lxml')
#soup.find('title').text # Заголовок
soup.find( class_='single-article__post-content single-article__content-container').text
```

Выгружаем заголовки и тесты статей

```python
base_url1='https://lifehacker.ru'
result = []
for url in parsed_urls:
    article = {}
    article_url = f'{base_url1}{url}'
    print(article_url)
    w=article_url
    p=w.find('https://lifehacker.ru/special')
    p1=w.find('https://burninghut.ru/')

    if p==-1 and p1==-1: # Исключаем сслыки на каталоги и внешние сайты
      response = requests.get(article_url)
      soup = BeautifulSoup(response.text, 'lxml')


      article['title'] = soup.find('title').text

      article['text'] = soup.find( class_='single-article__post-content single-article__content-container').text



      result.append(article)

```

Создаем dataframe

```python
import pandas as pd
pd.set_option('display.max_colwidth', 400)

data = pd.DataFrame(result)
data.head()
```
